```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(D.ExpectedMortality, bins=bins, color='#abcdef')
plt.hist(D.DischargeDisposition, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Expected Mortality')
plt.xlabel('DischargeDisposition')
plt.title('to show the distribution of expected mortality')

plt.savefig('histogram.png', dpi=300)

plt.show()
```
