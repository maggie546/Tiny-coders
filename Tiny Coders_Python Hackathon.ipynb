```python
#List male patients over the age of 60 who have undergone surgery from January-March 2019
d1 = datetime.datetime(1998, 1, 1)
d2 = datetime.datetime(2001, 3, 31)
agelimit=60
pmsl = pd.DataFrame(columns=['patSurgeryList'])
for (index, row) in P.iterrows():
    if row.ProviderSpecialty == 'Surgery' and (row.ProviderDateOnStaff >= d1) and (row.ProviderDateOnStaff <= d2): 
        name = row.ProviderName
        for index, rowPA in PA.iterrows():
            nameC = rowPA.FirstName+ ' ' +rowPA.LastName            
            agedelta = datetime.datetime.now() - rowPA.DateOfBirth
            if rowPA.Gender == 'Male' and name == nameC and (agedelta.days/365.25)<agelimit:
                pmsl=pmsl.append({'patSurgeryList': nameC}, ignore_index=True)
                break
pmsl        
```

```python
# Python code to reverse a string 
#Reverse the string without using inbuilt functions
# using loop
  
def reverse(s):
  str = ""
  for i in s:
    str = i + str
  return str
  
s = "PythonHackathon"
  
print ("The original string  is : ",end="")
print (s)
  
print ("The reversed string(using loops) is : ",end="")
print (reverse(s))
```

```python
# Get the count of number of patients based on ReasonForVisit and EDDisposition. Use stack/unstack functions

out = ED[['ReasonForVisit','EDDisposition']].value_counts().sort_index()
out

gen_unstack = out.unstack()
print(gen_unstack)
 

unstack_level1 = out.stack(level=0).unstack()
print(unstack_level1)
 

unstack_level2 = out.stack(level=1).unstack()
print(unstack_level2)
```
