```python
import numpy as np
import matplotlib.pyplot as plt
 
 
# Creating dataset
Team = ['Ashwini', 'Bhuvee', 'Manjushree',
        'Meghana', 'Leftovers']
 
data = [19, 18, 18, 19, 10]
 
 
# Creating explode data
explode = (0.0, 0.0, 0.0, 0.0, 0.2)
 
# Creating color parameters
colors = ( "orange", "cyan", "pink",
          "blue", "green")
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} ques)".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  
                                 autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = Team,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
 
# Adding legend
ax.legend(wedges, Team ,
          title ="Team ",
          loc ="center left",
          bbox_to_anchor =(1.2, 0, 0.5, 1))
 
plt.setp(autotexts, size = 10, weight ="bold")
ax.set_title("Showing our team's progress using pie chart")
 
# show plot
plt.show()
```
